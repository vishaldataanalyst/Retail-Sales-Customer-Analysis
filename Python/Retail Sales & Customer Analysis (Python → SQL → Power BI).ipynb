```python
rfm_sql = pd.read_sql("SELECT * FROM rfm_sql", con=engine)
```

```python
rfm_sql['R_Score'] = pd.qcut(rfm_sql['Recency'], 5, labels=range(5,0,-1))
rfm_sql['F_Score'] = pd.qcut(rfm_sql['Frequency'].rank(method='first'), 5, labels=range(1,6))
rfm_sql['M_Score'] = pd.qcut(rfm_sql['Monetary'].rank(method='first'), 5, labels=range(1,6))
rfm_sql['RFM_Score'] = rfm_sql['R_Score'].astype(str) + rfm_sql['F_Score'].astype(str) + rfm_sql['M_Score'].astype(str)
rfm_sql.head(10)
```

```python
rfm_sql['RFM_Score'].value_counts()
```

```python
rfm_sql.to_sql(name='rfm_customers', con=engine, if_exists='replace', index=False)
print("RFM table uploaded successfully!")
```
